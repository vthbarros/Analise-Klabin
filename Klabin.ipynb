```python
import pandas as pd
import numpy as np
import matplotlib as plt
import yfinance as yf
import mplfinance as mpf
```

### Extração dos Dados do Yahoo Finance

```python
klabin = yf.download("KLBN4.SA", period="max")
print(klabin)
```

```python
mpf.plot(klabin,
         type="line",
         mav=(3), #define a média móvel
         style="yahoo", 
         volume=True,
         title="Gráfico da Klabin (KLBN4.SA) em Candlestick usando Python",
         ylabel="Preços",
         ylabel_lower="Volume",
         xlabel="Data",
         figsize=(20,10)) #define o tamanho do gráfico gerado, sendo (Largura, Altura)
```
